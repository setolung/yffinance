```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

def plot_stock_price(days, ticker):
    # Set the start and end date for the historical data
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=days)).strftime('%Y-%m-%d')
    
    # Get the historical data from Yahoo Finance
    tickerData = yf.Ticker(ticker)
    df = tickerData.history(start=start_date, end=end_date)
    
    # Plot the closing prices in a line chart
    plt.plot(df['Close'])
    plt.title(ticker + ' Closing Prices (Last ' + str(days) + ' Days)')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()

# Example usage: plot the last 60 days of GBP/HKD prices
ticker = "GBPHKD=X"
plot_stock_price(90, ticker)
```

```python
data = yf.download(ticker, period="90d", interval="1d")

last_price = data.iloc[-1]["Close"]

# if it is bottom 20
bottom20_threshold = data["Close"].quantile(0.2)

if last_price <= bottom20_threshold:
    print("Buy it now")
else:
    print("Let's wait")
```
